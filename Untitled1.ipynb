```python
# Cell 1: imports & setup
import warnings, os, numpy as np, pandas as pd
warnings.filterwarnings("ignore")

import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, classification_report, roc_auc_score, roc_curve
)

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)

print("Setup complete ✅")
```

```python
# Cell 2: load data (use the exact filename you saw after upload)
# If your file name differs, paste it below exactly.
fname = "train_u6lujuX_CVtuZ9i (1).csv"

assert os.path.exists(fname), f"File not found: {fname}"
data = pd.read_csv(fname)
print("Shape:", data.shape)
data.head()
```

```python
# Cell 3: quick overview
print("Columns:", list(data.columns))
print("\nInfo:")
print(data.info())
print("\nMissing values per column:")
print(data.isnull().sum())
```

```python
# Cell 4: map target & drop ID
# Drop obvious ID column if present
for cid in ["Loan_ID", "ID", "id"]:
    if cid in data.columns:
        data = data.drop(columns=[cid])

# Map Y/N to 1/0 for Loan_Status
assert "Loan_Status" in data.columns, "Expected target column 'Loan_Status' not found."
data["Loan_Status"] = data["Loan_Status"].map({"Y": 1, "N": 0})

print("Target value counts:")
print(data["Loan_Status"].value_counts(dropna=False))
data.head()
```

```python
# Cell 5: split X / y and train/test
target_col = "Loan_Status"
X = data.drop(columns=[target_col])
y = data[target_col]

# Identify column types
categorical_cols = X.select_dtypes(include=["object"]).columns.tolist()
numeric_cols = X.select_dtypes(include=[np.number]).columns.tolist()

print("Categorical:", categorical_cols)
print("Numeric:", numeric_cols)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=RANDOM_STATE
)
print("Train:", X_train.shape, " Test:", X_test.shape)
```

```python
# Cell 6: preprocessing + model pipeline
numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("ohe", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_cols),
        ("cat", categorical_transformer, categorical_cols),
    ]
)

log_reg = LogisticRegression(
    max_iter=1000,
    # class_weight="balanced"  # ← uncomment if your classes are very imbalanced
)

model = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("clf", log_reg)
])

model.fit(X_train, y_train)
print("Model trained ✅")
```

```python
# Cell 7: metrics
y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)[:, 1]

acc  = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred, zero_division=0)
rec  = recall_score(y_test, y_pred, zero_division=0)
f1   = f1_score(y_test, y_pred, zero_division=0)
auc  = roc_auc_score(y_test, y_proba)

print(f"Accuracy : {acc:.3f}")
print(f"Precision: {prec:.3f}")
print(f"Recall   : {rec:.3f}")
print(f"F1-score : {f1:.3f}")
print(f"ROC AUC  : {auc:.3f}\n")

print("Classification Report:\n")
print(classification_report(y_test, y_pred, digits=3))
```

```python
# Cell 8: confusion matrix plot
cm = confusion_matrix(y_test, y_pred)

plt.figure()
plt.imshow(cm, interpolation="nearest")
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.xticks([0,1],[0,1])
plt.yticks([0,1],[0,1])

# annotate cells
for (i, j), z in np.ndenumerate(cm):
    plt.text(j, i, str(z), ha="center", va="center")

plt.colorbar()
plt.show()
```

```python
# Cell 9: ROC curve
fpr, tpr, thr = roc_curve(y_test, y_proba)

plt.figure()
plt.plot(fpr, tpr, label=f"AUC = {auc:.3f}")
plt.plot([0,1],[0,1],"--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend(loc="lower right")
plt.show()
```

```python
# Cell 10: map coefficients back to feature names
# Extract the trained pieces
prep = model.named_steps["preprocessor"]
clf  = model.named_steps["clf"]

# Get names after transformation
ohe = prep.named_transformers_["cat"].named_steps["ohe"]
cat_feature_names = ohe.get_feature_names_out(categorical_cols).tolist()
num_feature_names = numeric_cols
all_feature_names = num_feature_names + cat_feature_names

coef = clf.coef_.ravel()
coef_df = pd.DataFrame({"feature": all_feature_names, "coefficient": coef})
coef_df_sorted = coef_df.sort_values("coefficient", ascending=False)

print("Top positive features (increase approval odds):")
display(coef_df_sorted.head(12))

print("\nTop negative features (decrease approval odds):")
display(coef_df_sorted.tail(12))
```

```python
# Cell 11: single prediction demo using one row from test set
example = X_test.iloc[[0]].copy()
proba = model.predict_proba(example)[:, 1][0]
pred  = int(model.predict(example)[0])

print("Example prediction — probability of approval:", round(proba, 3))
print("Predicted class (1=Approved, 0=Rejected):", pred)
example
```
